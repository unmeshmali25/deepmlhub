## Part 7.1 Single Request

```python
import requests
import json
def check_ollama_status():
    """Check if Ollama server is running and accessible."""
    try:
        response = requests.get('http://localhost:11434/api/tags', timeout=5)
        if response.status_code == 200:
            models = response.json().get('models', [])
            print(f"✓ Ollama is running!")
            print(f"✓ Available models: {[m['name'] for m in models]}")
            return True
    except requests.exceptions.ConnectionError:
        print("✗ Ollama server not running")
        print("  Run: OLLAMA_NUM_PARALLEL=4 OLLAMA_CONTEXT_LENGTH=2048 ollama serve")
        return False
    except Exception as e:
        print(f"✗ Error: {e}")
        return False
```

```python
import time
import json
import requests
def test_single_request(model_name="qwen3.5:2b", prompt="Test prompt"):
    """
    Test a single synchronous request and measure timing.
    
    Returns:
        dict: Results with TTFT, total time, and response
    """
    
    # Prepare the request payload
    payload = {
        "model": model_name,
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        "stream": False,  # Wait for full response
#         "options": {
#             "temperature": 0.7,
#             "num_predict": 100  # Limit output tokens
#         }
    }
    
    print(f"\n{'='*60}")
    print(f"Testing single request with {model_name}")
    print(f"Prompt: {prompt[:50]}...")
    print(f"{'='*60}")
    
    # Measure total time
    start_time = time.time()
    
    try:
        # Send request
        response = requests.post(
            'http://localhost:11434/api/chat',
            json=payload,
            timeout=30
        )
        
        # Calculate timing
        total_time = time.time() - start_time
        
        if response.status_code == 200:
            result = response.json()
            content = result['message']['content']
            
            print(f"\n✓ Request successful!")
            print(f"  Total time: {total_time:.3f}s")
            print(f"  Response length: {len(content)} chars")
            print(f"  Response: {content[:100]}...")
            
            return {
                'success': True,
                'total_time': total_time,
                'content': content,
                'model': model_name
            }
        else:
            print(f"\n✗ Request failed: {response.status_code}")
            print(f"  Response: {response.text}")
            return {'success': False, 'error': response.text}
            
    except Exception as e:
        print(f"\n✗ Exception: {e}")
        return {'success': False, 'error': str(e)}
```

```python
for i in range(3): 
    result = test_single_request(
        model_name="qwen3.5:2b",
        prompt="What is 4+5?"
    )
```

## Part 7.2 Using asyncio

```python
import asyncio
import aiohttp
import time
from typing import List, Dict
class OllamaAsyncQueue:
    def __init__(self, max_concurrent: int = 4, timeout: int = 30, max_retries: int = 3, base_url: str = "http://localhost:11434"):
        self.max_concurrent, self.timeout, self.max_retries, self.base_url = max_concurrent, timeout, max_retries, base_url
        self.semaphore = asyncio.Semaphore(max_concurrent)
    
    async def _make_request(self, session: aiohttp.ClientSession, payload: Dict, request_id: int) -> Dict:
        url, start_time = f"{self.base_url}/api/chat", time.time()
        for attempt in range(self.max_retries):
            try:
                async with self.semaphore:
                    async with session.post(url, json=payload, timeout=aiohttp.ClientTimeout(total=self.timeout)) as response:
                        if response.status == 200:
                            result = await response.json()
                            return {'request_id': request_id, 'success': True, 'total_time': time.time() - start_time, 'content': result['message']['content']}
                        raise Exception(f"HTTP {response.status}")
            except Exception:
                if attempt < self.max_retries - 1:
                    await asyncio.sleep(2 ** attempt)
        return {'request_id': request_id, 'success': False, 'total_time': time.time() - start_time, 'error': 'Max retries exceeded'}
    
    async def process_requests(self, prompts: List[str], model_name: str = "qwen3.5:2b") -> Dict:
        start_time = time.time()
        async with aiohttp.ClientSession() as session:
            tasks = [self._make_request(session, {"model": model_name, "messages": [{"role": "user", "content": p}], "stream": False}, i) for i, p in enumerate(prompts)]
            results = await asyncio.gather(*tasks)
        total_time = time.time() - start_time
        successful = [r for r in results if r['success']]
        return {
            'total_requests': len(prompts),
            'successful': len(successful),
            'failed': len(prompts) - len(successful),
            'total_time': total_time,
            'throughput': len(prompts) / total_time,
            'avg_time_per_request': sum(r['total_time'] for r in successful) / len(successful) if successful else 0,
            'results': results
        }
async def test_async_queue():
    queue = OllamaAsyncQueue(max_concurrent=4)
    prompts = [f"What is {i}+{i}?" for i in range(5)]
    return await queue.process_requests(prompts)
```

```python
await test_async_queue()
```

## Part 7.3 Batching

```python
import asyncio
import aiohttp
import time
from typing import List, Dict, Callable
import ollama
class FixedBatchProcessor:
    """
    Fixed-Size Batching Strategy
    
    Idea: Process requests in fixed-size chunks (e.g., 2 at a time)
    
    Mental Model: Like filling a bus - wait until you have N passengers,
    then drive. Next batch waits for the bus to return.
    """
    
    def __init__(self, batch_size: int = 2, base_url: str = "http://localhost:11434"):
        self.batch_size = batch_size
        self.base_url = base_url
    
    async def _send_single(self, session: aiohttp.ClientSession, 
                          prompt: str, request_id: int, model: str) -> Dict:
        """Send one request and measure timing"""
        start = time.time()
        payload = {
            "model": model,
            "messages": [{"role": "user", "content": prompt}],
            "stream": False
        }
        
        try:
            async with session.post(
                f"{self.base_url}/api/chat", 
                json=payload,
                timeout=aiohttp.ClientTimeout(total=60)
            ) as resp:
                if resp.status == 200:
                    data = await resp.json()
                    return {
                        'request_id': request_id,
                        'success': True,
                        'latency': time.time() - start,
                        'content': data['message']['content'][:50]  # Truncate for display
                    }
        except Exception as e:
            return {
                'request_id': request_id,
                'success': False,
                'latency': time.time() - start,
                'error': str(e)
            }
    
    async def process_batch(self, prompts: List[str], model: str = "qwen3.5:2b") -> Dict:
        """
        Process prompts in fixed-size batches
        
        Example with batch_size=2, 5 prompts:
        - Batch 1: [0, 1] -> wait -> Batch 2: [2, 3] -> wait -> Batch 3: [4]
        """
        all_results = []
        total_start = time.time()
        
        # Split into batches
        batches = [
            prompts[i:i + self.batch_size] 
            for i in range(0, len(prompts), self.batch_size)
        ]
        
        print(f"Processing {len(prompts)} requests in {len(batches)} batches "
              f"(batch_size={self.batch_size})")
        
        async with aiohttp.ClientSession() as session:
            for batch_num, batch_prompts in enumerate(batches):
                batch_start = time.time()
                
                # Send all requests in this batch concurrently
                # Note: Even though we use asyncio.gather, Ollama will process
                # them based on OLLAMA_NUM_PARALLEL setting
                tasks = [
                    self._send_single(session, prompt, idx + batch_num * self.batch_size, model)
                    for idx, prompt in enumerate(batch_prompts)
                ]
                
                batch_results = await asyncio.gather(*tasks)
                batch_time = time.time() - batch_start
                
                all_results.extend(batch_results)
                print(f"  Batch {batch_num + 1}/{len(batches)}: "
                      f"{len(batch_results)} requests in {batch_time:.2f}s")
        
        total_time = time.time() - total_start
        
        successful = [r for r in all_results if r['success']]
        return {
            'strategy': 'fixed_batch',
            'batch_size': self.batch_size,
            'total_requests': len(prompts),
            'successful': len(successful),
            'total_time': total_time,
            'throughput': len(prompts) / total_time,
            'avg_latency': sum(r['latency'] for r in successful) / len(successful) if successful else 0,
            'results': all_results
        }
# Test it
async def test_fixed_batching():
    # Create varied prompts (simulating different agents)
    prompts = [
        f"Agent {i}: Should I shop today? Budget=${50+i*10}, Day=Mon" 
        for i in range(6)
    ]
    
    # Test with batch_size=2
    processor = FixedBatchProcessor(batch_size=2)
    return await processor.process_batch(prompts, model="qwen3.5:2b")
```

```python
await test_fixed_batching()
```
