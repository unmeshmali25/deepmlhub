```python
import requests
import json
def check_ollama_status():
    """Check if Ollama server is running and accessible."""
    try:
        response = requests.get('http://localhost:11434/api/tags', timeout=5)
        if response.status_code == 200:
            models = response.json().get('models', [])
            print(f"✓ Ollama is running!")
            print(f"✓ Available models: {[m['name'] for m in models]}")
            return True
    except requests.exceptions.ConnectionError:
        print("✗ Ollama server not running")
        print("  Run: OLLAMA_NUM_PARALLEL=4 OLLAMA_CONTEXT_LENGTH=2048 ollama serve")
        return False
    except Exception as e:
        print(f"✗ Error: {e}")
        return False
```

```python
import time
import json
import requests
def test_single_request(model_name="qwen3.5:2b", prompt="Test prompt"):
    """
    Test a single synchronous request and measure timing.
    
    Returns:
        dict: Results with TTFT, total time, and response
    """
    
    # Prepare the request payload
    payload = {
        "model": model_name,
        "messages": [
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ],
        "stream": False,  # Wait for full response
#         "options": {
#             "temperature": 0.7,
#             "num_predict": 100  # Limit output tokens
#         }
    }
    
    print(f"\n{'='*60}")
    print(f"Testing single request with {model_name}")
    print(f"Prompt: {prompt[:50]}...")
    print(f"{'='*60}")
    
    # Measure total time
    start_time = time.time()
    
    try:
        # Send request
        response = requests.post(
            'http://localhost:11434/api/chat',
            json=payload,
            timeout=30
        )
        
        # Calculate timing
        total_time = time.time() - start_time
        
        if response.status_code == 200:
            result = response.json()
            content = result['message']['content']
            
            print(f"\n✓ Request successful!")
            print(f"  Total time: {total_time:.3f}s")
            print(f"  Response length: {len(content)} chars")
            print(f"  Response: {content[:100]}...")
            
            return {
                'success': True,
                'total_time': total_time,
                'content': content,
                'model': model_name
            }
        else:
            print(f"\n✗ Request failed: {response.status_code}")
            print(f"  Response: {response.text}")
            return {'success': False, 'error': response.text}
            
    except Exception as e:
        print(f"\n✗ Exception: {e}")
        return {'success': False, 'error': str(e)}
```

## Part 7.2 Using asyncio

```python
import asyncio
import aiohttp
import time
from typing import List, Dict
class OllamaAsyncQueue:
    def __init__(self, max_concurrent: int = 4, timeout: int = 30, max_retries: int = 3, base_url: str = "http://localhost:11434"):
        self.max_concurrent, self.timeout, self.max_retries, self.base_url = max_concurrent, timeout, max_retries, base_url
        self.semaphore = asyncio.Semaphore(max_concurrent)
    
    async def _make_request(self, session: aiohttp.ClientSession, payload: Dict, request_id: int) -> Dict:
        url, start_time = f"{self.base_url}/api/chat", time.time()
        for attempt in range(self.max_retries):
            try:
                async with self.semaphore:
                    async with session.post(url, json=payload, timeout=aiohttp.ClientTimeout(total=self.timeout)) as response:
                        if response.status == 200:
                            result = await response.json()
                            return {'request_id': request_id, 'success': True, 'total_time': time.time() - start_time, 'content': result['message']['content']}
                        raise Exception(f"HTTP {response.status}")
            except Exception:
                if attempt < self.max_retries - 1:
                    await asyncio.sleep(2 ** attempt)
        return {'request_id': request_id, 'success': False, 'total_time': time.time() - start_time, 'error': 'Max retries exceeded'}
    
    async def process_requests(self, prompts: List[str], model_name: str = "qwen3.5:2b") -> Dict:
        start_time = time.time()
        async with aiohttp.ClientSession() as session:
            tasks = [self._make_request(session, {"model": model_name, "messages": [{"role": "user", "content": p}], "stream": False}, i) for i, p in enumerate(prompts)]
            results = await asyncio.gather(*tasks)
        total_time = time.time() - start_time
        successful = [r for r in results if r['success']]
        return {
            'total_requests': len(prompts),
            'successful': len(successful),
            'failed': len(prompts) - len(successful),
            'total_time': total_time,
            'throughput': len(prompts) / total_time,
            'avg_time_per_request': sum(r['total_time'] for r in successful) / len(successful) if successful else 0,
            'results': results
        }
async def test_async_queue():
    queue = OllamaAsyncQueue(max_concurrent=4)
    prompts = [f"What is {i}+{i}?" for i in range(5)]
    return await queue.process_requests(prompts)
```

## Part 7.3 Batching

```python
import time
from typing import List, Dict
import ollama
class SimpleBatchProcessor:
    """
    Dead simple: create client once, loop through prompts
    """
    
    def __init__(self, base_url: str = "http://localhost:11434"):
        self.client = ollama.Client(host=base_url)
    
    def process_batch(self, prompts: List[str], model: str = "qwen3.5:2b") -> Dict:
        results = []
        total_start = time.time()
        
        for i, prompt in enumerate(prompts):
            start = time.time()
            
            try:
                resp = self.client.generate(model=model, prompt=prompt)
                elapsed = time.time() - start
                
                results.append({
                    'request_id': i,
                    'success': True,
                    'latency': elapsed,
                    'content': resp.get('response', '')[:100]
                })
                
                print(f"  [{i+1}/{len(prompts)}] Done in {elapsed:.2f}s")
                
            except Exception as e:
                elapsed = time.time() - start
                results.append({
                    'request_id': i,
                    'success': False,
                    'latency': elapsed,
                    'error': str(e)
                })
                print(f"  [{i+1}/{len(prompts)}] Error: {e}")
        
        total_time = time.time() - total_start
        
        return {
            'total_requests': len(prompts),
            'total_time': total_time,
            'throughput': len(prompts) / total_time,
            'results': results
        }
# Test it
prompts = [
    f"Agent {i}: What is {i} + {i+1}?" 
    for i in range(4)
]
processor = SimpleBatchProcessor()
print("Starting batch...")
results = processor.process_batch(prompts, model="qwen3.5:2b")
print(f"\nTotal: {results['total_time']:.2f}s ({results['throughput']:.2f} req/s)")
```

```python
import time
from typing import List, Dict
import ollama
class SimpleBatchProcessor:
    def __init__(self, base_url: str = "http://localhost:11434"):
        self.client = ollama.Client(host=base_url)
    
    def process_batch(self, prompts: List[str], model: str = "qwen3.5:2b") -> Dict:
        results = []
        total_start = time.time()
        
        for i, prompt in enumerate(prompts):
            req_start = time.time()
            
            try:
                # NO OPTIONS - use Ollama defaults
                resp = self.client.generate(model=model, prompt=prompt)
                elapsed = time.time() - req_start
                
                # Truncate for display
                display = resp.response[:60]
                if len(resp.response) > 60:
                    display += "..."
                
                results.append({
                    'request_id': i,
                    'success': True,
                    'latency': elapsed,
                    'content': resp.response,
                    'tokens': resp.eval_count,
                    'prompt': prompt[:40]
                })
                
                print(f"[{i+1}/{len(prompts)}] {elapsed:.2f}s ({resp.eval_count}tok): {display}")
                
            except Exception as e:
                elapsed = time.time() - req_start
                results.append({
                    'request_id': i,
                    'success': False,
                    'latency': elapsed,
                    'error': str(e),
                    'prompt': prompt[:40]
                })
                print(f"[{i+1}/{len(prompts)}] ERROR: {e}")
        
        total_time = time.time() - total_start
        successful = [r for r in results if r['success']]
        
        return {
            'total_requests': len(prompts),
            'successful': len(successful),
            'total_time': total_time,
            'avg_tokens': sum(r['tokens'] for r in successful) / len(successful) if successful else 0,
            'throughput': len(prompts) / total_time,
            'results': results
        }
# Test with 4 prompts
prompts = [
    "What is 4+5?",
    "Explain quantum computing simply",
    "Name three colors",
    "What day comes after Tuesday?"
]
processor = SimpleBatchProcessor()
print("Running batch...\n")
results = processor.process_batch(prompts, model="qwen3.5:2b")
print(f"\n{'='*60}")
print(f"Completed: {results['successful']}/{results['total_requests']}")
print(f"Total time: {results['total_time']:.2f}s")
print(f"Throughput: {results['throughput']:.2f} req/s")
print(f"Avg tokens: {results['avg_tokens']:.0f}")
print(f"{'='*60}")
```

```python
import time
import asyncio
from typing import List, Dict
import ollama
class LLMBenchmark:
    def __init__(self, model="qwen3.5:2b"):
        self.model = model
        self.client = ollama.Client()
    
    def run_sequential(self, prompts: List[str]) -> Dict:
        """One at a time"""
        print(f"\n{'='*60}")
        print("SEQUENTIAL (one at a time)")
        print(f"{'='*60}")
        
        results = []
        start_total = time.time()
        
        for i, prompt in enumerate(prompts):
            start = time.time()
            resp = self.client.generate(model=self.model, prompt=prompt)
            elapsed = time.time() - start
            
            display = resp.response[:30] + "..." if len(resp.response) > 30 else resp.response
            results.append({
                'id': i,
                'latency': elapsed,
                'tokens': resp.eval_count,
                'response': display
            })
            print(f"  [{i+1}/{len(prompts)}] {elapsed:.2f}s ({resp.eval_count}tok): {display}")
        
        total = time.time() - start_total
        return {
            'strategy': 'sequential',
            'total_time': total,
            'throughput': len(prompts) / total,
            'avg_latency': sum(r['latency'] for r in results) / len(results),
        }
    
    async def _async_one(self, prompt: str, idx: int) -> Dict:
        start = time.time()
        
        loop = asyncio.get_event_loop()
        resp = await loop.run_in_executor(
            None,
            lambda: self.client.generate(model=self.model, prompt=prompt)
        )
        
        return {
            'id': idx,
            'latency': time.time() - start,
            'tokens': resp.eval_count
        }
    
    async def run_async_all(self, prompts: List[str]) -> Dict:
        """All at once"""
        print(f"\n{'='*60}")
        print(f"ASYNC CONCURRENT (all {len(prompts)} at once)")
        print(f"{'='*60}")
        
        start_total = time.time()
        tasks = [self._async_one(p, i) for i, p in enumerate(prompts)]
        results = await asyncio.gather(*tasks)
        total = time.time() - start_total
        
        for r in results:
            print(f"  [{r['id']+1}/{len(prompts)}] {r['latency']:.2f}s ({r['tokens']}tok)")
        
        return {
            'strategy': 'async_concurrent',
            'total_time': total,
            'throughput': len(prompts) / total,
            'avg_latency': sum(r['latency'] for r in results) / len(results),
        }
    
    async def run_batch(self, prompts: List[str], batch_size: int = 2) -> Dict:
        """Fixed-size batches"""
        print(f"\n{'='*60}")
        print(f"FIXED BATCHING (size={batch_size})")
        print(f"{'='*60}")
        
        results = []
        start_total = time.time()
        
        batches = [prompts[i:i + batch_size] for i in range(0, len(prompts), batch_size)]
        
        for batch_num, batch in enumerate(batches):
            batch_start = time.time()
            tasks = [self._async_one(p, i + batch_num*batch_size) for i, p in enumerate(batch)]
            batch_results = await asyncio.gather(*tasks)
            batch_time = time.time() - batch_start
            
            results.extend(batch_results)
            print(f"  Batch {batch_num+1}: {len(batch)} requests in {batch_time:.2f}s")
        
        total = time.time() - start_total
        return {
            'strategy': f'batch_{batch_size}',
            'total_time': total,
            'throughput': len(prompts) / total,
            'avg_latency': sum(r['latency'] for r in results) / len(results),
        }
# Generate math prompts
prompts = [f"What is 2+{i}?" for i in range(2, 8)]  # 2+2, 2+3, 2+4, 2+5, 2+6, 2+7
print(f"Testing with {len(prompts)} math prompts:")
for i, p in enumerate(prompts):
    print(f"  {i+1}. {p}")
benchmark = LLMBenchmark()
results = []
# Run all strategies
results.append(benchmark.run_sequential(prompts))
results.append(await benchmark.run_async_all(prompts))
results.append(await benchmark.run_batch(prompts, batch_size=2))
results.append(await benchmark.run_batch(prompts, batch_size=3))
# Summary
print(f"\n{'='*60}")
print("BENCHMARK SUMMARY")
print(f"{'='*60}")
print(f"{'Strategy':<20} {'Total':<10} {'Throughput':<12} {'Avg Latency':<12}")
print("-" * 60)
for r in sorted(results, key=lambda x: x['throughput'], reverse=True):
    print(f"{r['strategy']:<20} {r['total_time']:>6.2f}s    {r['throughput']:>6.2f} req/s    {r['avg_latency']:>6.2f}s")
```
